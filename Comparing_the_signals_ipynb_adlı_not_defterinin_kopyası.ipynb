```python
import numpy as np
import matplotlib.pyplot as plt

# Generate two time-domain signals
signal1 = np.sin(np.linspace(0, 10, 100))
signal2 = np.sin(np.linspace(0, 10, 100) + np.pi/4)

# Calculate the cross-correlation
correlation = np.correlate(signal1, signal2, mode='full')

# Plot the first signal
plt.figure()
plt.plot(signal1)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Signal 1")

# Plot the second signal
plt.figure()
plt.plot(signal2)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Signal 2")

# Plot the cross-correlation function
plt.figure()
plt.plot(correlation)
plt.xlabel("Lag")
plt.ylabel("Correlation")
plt.title("Cross-correlation")

plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Generate two time-domain signals
signal1 = np.sin(np.linspace(0, 10, 100))
signal2 = np.sin(np.linspace(0, 10, 100) + np.pi/4)

# Perform the Fourier transform
fourier1 = np.fft.fft(signal1)
fourier2 = np.fft.fft(signal2)

# Get the frequency axis
frequency = np.fft.fftfreq(signal1.shape[-1])

# Plot the first signal
plt.figure()
plt.plot(signal1)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Signal 1")

# Plot the second signal
plt.figure()
plt.plot(signal2)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Signal 2")

# Plot the frequency spectrums
plt.figure()
plt.plot(frequency, np.abs(fourier1))
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.title("Frequency Spectrum of signal 1")

plt.figure()
plt.plot(frequency, np.abs(fourier2))
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.title("Frequency Spectrum of signal 2")
```

Here are a few metrics that can be used to measure the similarity of non-linear signals:

**Dynamic Time Warping (DTW) distance:** DTW is a method for comparing two signals by warping the time axis of one signal to align with the other. It finds the optimal alignment between the two signals by minimizing the distance between the two time series. The resulting DTW distance is a measure of the similarity between the two signals.

**Frchet distance:** This is a measure of the similarity between two curves based on the minimum amount of deformation required to transform one curve into the other. It is calculated by finding the shortest path between the two curves.

**Hausdorff distance: **This is a measure of the similarity between two sets of points based on the maximum distance between any two points in the two sets.

**Earth Mover's Distance (EMD):** This is a measure of the distance between two probability distributions. It is calculated by finding the minimum amount of "work" required to transform one distribution into the other.

**Cosine Similarity:** This method compares the similarity of two signals by measuring the cosine of the angle between their vectors. The cosine similarity of two signals is calculated by taking the dot product of the two signals and dividing it by the product of their magnitudes.

**Dynamic Mode Decomposition (DMD): ** This method decomposes a time series into a set of spatial-temporal patterns and compares the modes of oscillation of the two time series.

```python
import numpy as np
from scipy.spatial.distance import cdist

# Generate two time-domain signals
signal1 = np.sin(np.linspace(0, 10, 100))
signal2 = np.sin(np.linspace(0, 10, 100) + np.pi/4)

# Calculate the DTW distance
distance, _, _, _ = cdist(np.array([signal1]), np.array([signal2]), metric='dtw')
#distance = cdist(np.array([signal1]), np.array([signal2]), metric='dtw')
print("DTW distance: ", distance[0])
```
